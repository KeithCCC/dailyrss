```python
import os
import pandas as pd
from tkinter import *

# Load the DataFrame
filename = "default.json"
if os.path.exists(filename):
    try:
        df = pd.read_json(filename)
        # Reorder the columns
        df = df[['label', 'title', 'url']]
    except Exception as e:
        print(e)

# Create the Tkinter window
root = Tk()
root.title("DataFrame Display")

# Create a Text widget
text = Text(root, wrap=NONE)
text.pack(expand=YES, fill=BOTH)

# Define a custom formatter to left-align all columns
def left_align_formatter(df):
    formatters = {col: '{{:<{}}}'.format(df[col].astype(str).str.len().max()).format for col in df.columns}
    return df.to_string(formatters=formatters, index=False)

# Insert DataFrame content into the Text widget with left alignment
text.insert(END, left_align_formatter(df))

# Run the Tkinter event loop
root.mainloop()
```
